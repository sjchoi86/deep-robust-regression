# Boston housing dataset & diabetes dataset

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
%config InlineBackend.figure_format = 'retina'
from sklearn.datasets import load_boston,load_diabetes
from util import nzr,get_train_test_datasets
print ("Package loaded.")
```

```python
def plot_dataset(x_train,x_test,y_train,y_test,dataset,feat_idx):
    feat_idx = 5 # select the feature index to plot 
    plt.figure(figsize=(10,4))
    h_tr,=plt.plot(x_train[:,feat_idx],y_train,'ko') # plot train data
    h_te,=plt.plot(x_test[:,feat_idx],y_test,'ro') # plot test data
    plt.xlabel(dataset.feature_names[feat_idx],fontsize=15)
    plt.ylabel('Target',fontsize=15)
    plt.title(dataset_name,fontsize=20)
    plt.legend([h_tr,h_te],['Train data','Test data'],fontsize=15)
    plt.show()
```

### Load boston housing dataset

```python
dataset,dataset_name = load_boston(),'boston'
x_train,x_test,y_train,y_test = get_train_test_datasets(
    dataset,dataset_name,_test_size=0.333,_outlier_rate=0.0,_SAVE_MAT=1,_VERBOSE=1)
feat_idx = 5
plot_dataset(x_train,x_test,y_train,y_test,dataset,feat_idx)
```

### Load boston housing with outliers

```python
dataset,dataset_name = load_boston(),'boston'
x_train,x_test,y_train,y_test = get_train_test_datasets(
    dataset,dataset_name,_test_size=0.333,_outlier_rate=0.5,_SAVE_MAT=0,_VERBOSE=0)
feat_idx = 5
plot_dataset(x_train,x_test,y_train,y_test,dataset,feat_idx)
```

### Load diabetes dataset

```python
dataset,dataset_name = load_diabetes(),'diabetes'
x_train,x_test,y_train,y_test = get_train_test_datasets(
    dataset,dataset_name,_test_size=0.333,_SAVE_MAT=True,_VERBOSE=True)
feat_idx = 8
plot_dataset(x_train,x_test,y_train,y_test,dataset,feat_idx)
```
